### 3.2.2 PCR in Practice
---

As derived in subsection $2.1.6$, it is assumed in this subsection that the true $\pmb C_X$ is unknown and therefore the matrix of eigenvectors $\pmb \phi$ is estimated by $\hat{\pmb \phi}$ prior the regression. I will denote $\beta_s$ with the s-index to indicate the stochastic case.

\begin{align}
Y =  \hat{\pmb Z} \beta_s + \varepsilon_Z = \pmb X \hat{\pmb \phi} \beta_s + \varepsilon_Z,
\tag{3.18}
\end{align}

The estimate is given analogously to equation *(2.16)* by

\begin{align}
\hat{\beta}_s = \arg \min\limits_{b \in \mathbb{R}^M} ||Y- \hat{\pmb Z} b||^2_2 = \left (\hat{\pmb Z'} \hat{\pmb Z} \right)^{-1} \hat{\pmb Z'} Y = \left (\hat{\pmb \phi'} \pmb X' \pmb X \hat{\pmb \phi} \right)^{-1}\hat{\pmb \phi'} \pmb X' Y.
\tag{3.19}
\end{align}

First, I will derive that the estimate conditioned on $\hat{\pmb Z}$ yields the true parameter. This finding indicates, using the law of iterated expectations, that the unconditional parameter is unbiased. Hence, the conditional variance of $\hat{\pmb  Z}$ can be computed as in the ordinary OLS case yielding $\text{Var}(\hat{\beta}_s|\pmb Z) = \sigma^2_z \left(\hat{\pmb Z'} \hat{\pmb Z} \right)^{-1}$. The latter three findings are subsequent used to compute the variance of $\hat{\pmb  Z}$, making use of the variance decomposition in the same fashion as in equation *(2.17)*.

\begin{align}
\text{E} \left(  \hat{\beta_s} \mid \hat{\pmb Z} \right) = \text{E} \left[ \left( \hat{\pmb Z'} \hat{\pmb Z} \right)^{-1}\hat{\pmb Z'} Y  \mid \hat{\pmb Z} \right] = \text{E} \left[ \left( \hat{\pmb Z'} \hat{\pmb Z} \right)^{-1}\hat{\pmb Z'} (\hat{\pmb Z}\beta_Z^s + \varepsilon_Z)  \mid \hat{\pmb Z} \right] 
= \beta_s +  \text{E} \left[ \left( \hat{\pmb Z'} \hat{\pmb Z} \right)^{-1}\hat{\pmb Z'} \varepsilon_Z \mid \hat{\pmb Z}  \right]  = \beta_s + \left( \hat{\pmb Z'} \hat{\pmb Z} \right)^{-1}\hat{\pmb Z'} \underbrace{\text{E}  \left[\varepsilon_Z \mid \hat{\pmb Z}  \right]}_{= 0} = \beta_s
\tag{3.20}
\end{align}

\begin{align}
\text{Var}(\hat{\beta}_s) \stackrel{\text{i}}{=} \text{E} \left[\text{Var}(\hat{\beta_s}|\hat{\pmb Z}) \right] + \text{Var} \left[\text{E}(\hat{\beta_s}| \hat{\pmb Z}) \right] \stackrel{\text{ii}}{=} \text{E} \left[\text{Var}(\hat{\beta_s}|\hat{\pmb Z}) \right] \stackrel{\text{iii}}{=} \text{E} \left[\hat{\sigma}^2_z  \left(\hat{\pmb Z'} \hat{\pmb Z} \right)^{-1} \right] = \text{E} \left[\hat{\sigma}^2_z \left(\hat{\pmb \phi'} \pmb X' \pmb X \hat{\pmb \phi} \right)^{-1} \right] = \text{E} \left[ \hat{\sigma}^2_z \begin{pmatrix} 
    \hat{\lambda}_1 & \dots & 0\\
    \vdots & \vdots & \vdots \\
    0 & \dots & \hat{\lambda}_M
\end{pmatrix}^{-1} \right] = \text{E} \left[\hat{\sigma}^2_z \begin{pmatrix} 
    \hat{\lambda}_1^{-1} & \dots & 0\\
    \vdots & \vdots & \vdots \\
    0 & \dots & \hat{\lambda}_M^{-1}
\end{pmatrix} \right]
\tag{3.21}
\end{align}
and therefore neither the variance nor the expected value are variant to the choice of the representant in $[\hat{\phi}]$ if the representation in *(2.21)* is chosen.

## 3.3 Estimate the Variance of $\hat{\beta}$ in a Simulation Study
---

If the notation $\hat{\beta}$ is used in this section, this indicates that the given statement holds for for $\hat{\beta_s}$ and $\hat{\beta_t}$. 

### 3.3.1 Use the Empirical Distribution of $\hat{\beta}$ to Estimate the Variance of $\hat{\beta}$ in a Simulation Study

For this approach I fix the sample size $n_s$ and simulate a specified number of iterations $I$ to get different samples yielding $I$ results for $\hat{\beta}$ denoted by $\hat{\beta^{(i)}} \quad i = 1, \dots, I$ respectively. This beta sample serves as an estimate of the empirical distribution of $\hat{\beta}$. Denoting the simulation sample mean of $\hat{\beta^{(i)}}$ by $\overline{\hat{\beta}}$, an estimate of the variance $\text{Var}(\hat{\beta})$ can therefore be obtained by 

\begin{align}
  \widehat{\text{Var}(\hat{\beta})}^E_{n_s} = \frac{1}{I - 1} \sum_{i = 1}^I \left(\hat{\beta^{(i)}} - \overline{\hat{\beta}} \right)^2
  \tag{3.22}
\end{align}

$ \widehat{\text{Var}(\hat{\beta})}^E_{n_s}$ denotes the estimate of the variance of $\hat{\beta}$ using the empirical distribution (E) and sample sizes of $n_s$. A problem that arises using this technique for the case of the estimated eigenvectors is that $\hat{\beta^{(i)}}$ is dependent on the choice of $\hat{\pmb \phi}$ and therefore the coefficients vary more than they would do if the representation of $[\hat{\phi}]$ would be unique and thus the variance of the coefficient should increase. This issue is not faced in the case using the true matrix of eigenvectors, since the representant of $[\phi]$ is chosen prior to the simulation of the samples. However, this issue can be circumvent by making use of formula *(3.21)*, which yields an estimate of the variance of $\hat{\beta}_s$ that is only dependent on the eigenvalues corresponding to $[\hat{\phi}]$ and thus is invariant to the choice of $\hat{\pmb \phi}$.

### 3.3.2 Use Formula *(3.16)*/*(3.21)* to Estimate the Variance of $\hat{\beta}$ in a Simulation Study

As stated above, the non-uniqueness of the matrix of eigenvectors in the practical case raises a need to estimate the variance of the coefficient differently. For completness I hold the subsection more general and show not only that that the variance of $\hat{\beta_s}$ can be estimated in a simulation by using formula *(3.21)* but also that the variance of $\hat{\beta_t}$ can be estimated in a simulation by using formula *(3.16)*. Let $I$ again denote the number of iterations, $n_s$ a fixed sample size, $\hat{\lambda}_k^{(i)}$ the k-th eigenvalue at iteration $i$, $X^{(i)}$ the matrix of regressors at iteration $i$ and $(\hat{\sigma}^2_z)^{(i)}$ the estimated error variance at iteration $i$. According to the stated formulas the variance of $\hat{\beta}$ can be estimated in a simulation study by computing the simulation sample averages

\begin{align}
    \widehat{\text{Var}(\hat{\beta_s})}^F_{n_s} = \frac{1}{I} \sum _{i = 1}^I (\hat{\sigma}^2_z)^{(i)} \begin{pmatrix} 
    \frac{1}{\hat{\lambda}_1^{(i)}} & \dots & 0\\
    \vdots & \vdots & \vdots \\
    0 & \dots & \frac{1}{\hat{\lambda}_M^{(i)}}
\end{pmatrix} 
\tag{3.23}
\end{align}

\begin{align}
\widehat{\text{Var}(\hat{\beta_t})}^F_{n_s} =  \frac{1}{I} \sum _{i = 1}^I  (\hat{\sigma}^2_z)^{(i)} \left( \pmb \phi' \pmb X'^{(i)} \pmb X^{(i)} \pmb \phi \right)^{-1} 
\tag{3.24}
\end{align}

Summed up, it is possible to compute the variance of $\text{Var}(\hat{\beta_t}$ with *(3.22)* and *(3.24)* without any problems induced by the non-uniquness of $\pmb \phi$. However, the variance of $\text{Var}(\hat{\beta_s}$ is dependent on the choice of $\hat{\pmb \phi}$ and thus the empirical distribution would yield biased uninterpretable results. In Section *4.4* I analyse how much the latter effect is.
